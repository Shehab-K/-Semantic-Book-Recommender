```python
category_mapping={"Fiction":"Fiction",
                  "Juvenile Fiction":"Children's Fiction",
                  "Biography & Autobiography":"Nonfiction",
                  "History":"Nonfiction",
                  "Literary Criticism":"Nonfiction",
                  "Philosophy":"Nonfiction",
                  "Religion ":"Nonfiction",
                  "Comics & Graphic Novels":"Fiction",
                  "Drama":"Fiction",
                  "Juvenile Nonfiction":"Children's Nonfiction",
                  "Science":"Nonfiction",
                  "Poetry":"Fiction"
}

books["simple_categories"]=books["categories"].map(category_mapping)
```

```python
from transformers import pipeline

fiction_categories=["Fiction","Nonfiction"]

pipe = pipeline("zero-shot-classification", model="MoritzLaurer/deberta-v3-xsmall-zeroshot-v1.1-all-33")
```

```python
seq=books.loc[books["simple_categories"]=="Fiction","description"].reset_index(drop=True)[0]
```

```python
pipe(seq,fiction_categories)
```

```python
import numpy as np
def generate_prediction(seq,fiction_categories):
    pred= pipe(seq,fiction_categories)
    max_index=np.argmax(pipe(seq,fiction_categories)["scores"])
    max_label=pipe(seq,fiction_categories)["labels"][max_index]
    return max_label
```

```python
from tqdm import tqdm

actual_cats=[]
pred_cats=[]

for i in tqdm(range(0,300)):
    sequence=books.loc[books["simple_categories"]=="Fiction","description"].reset_index(drop=True)[i]
    pred_cats+=[generate_prediction(sequence,fiction_categories)]
    actual_cats+=["Fiction"]
```

```python
for i in tqdm(range(0,300)):
    sequence=books.loc[books["simple_categories"]=="Nonfiction","description"].reset_index(drop=True)[i]
    pred_cats+=[generate_prediction(sequence,fiction_categories)]
    actual_cats+=["Nonfiction"]
```

```python
predictions_df=pd.DataFrame({"actual_categories":actual_cats,"predicted_categories":pred_cats})

predictions_df
```

```python
predictions_df["correct_predictions"]=(np.where(predictions_df["actual_categories"]==predictions_df["predicted_categories"],1,0))
```

```python
predictions_df["correct_predictions"].sum()/len(predictions_df)
```

```python
isbn=[]
predicted_cats=[]

missing_cats=books.loc[books["simple_categories"].isna(),["isbn13","description"]].reset_index(drop=True)
```

```python
for i in tqdm(range(0,len(missing_cats))):
    sequence=missing_cats["description"][i]

    predicted_cats+=[generate_prediction(sequence,fiction_categories)]

    isbn+=[missing_cats["isbn13"][i]]
```

```python
missing_predicted_df=pd.DataFrame({"isbn13":isbn,"predicted_categories":predicted_cats})
```

```python
missing_predicted_df
```

```python
books=pd.merge(books,missing_predicted_df,on="isbn13",how="left")
books["simple_categories"]=np.where(books["simple_categories"].isna(),books["predicted_categories"],
 books["simple_categories"])
books=books.drop(columns=["predicted_categories"])
```
